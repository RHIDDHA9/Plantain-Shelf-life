```python
import numpy as np
import cv2
import os
import random
import matplotlib.pyplot as plt          
```

```python
DIRECTORY= r"/home/rhidhdha/Desktop/Plantain_Shelf_Life(Jadavpur_University)/plantain-test"
CATAGORIES= ['day0','day1','day2','day3','day4','day5','day6','day7','day8','day9','day12','day13','day14','day15','day16','day17','day18','day19','day20','day21','day22','day23','day24','day25','day26','day27','day28','day29','day30','day31','day32','day33','day34','day35','day36','day37','day38','day39','day40','day41','day42']
```

```python
data=[]

for categories in CATAGORIES:
    folder=os.path.join(DIRECTORY,categories)
    label=CATAGORIES.index(categories)
    
    
    for img in os.listdir(folder):
        img=os.path.join(folder,img)
        img_arr=cv2.imread(img)
        img_arr=cv2.resize(img_arr,(100,100))
        
        data.append([img_arr,label])
```

```python

random.shuffle(data)
```

```python
x=[]
y=[]

for features,label in data:
    x.append(features)
    y.append(label)

X= np.array(x)
Y=np.array(y)
```

```python
X=X/255
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense,Activation,Dropout
```

```python
model=Sequential()
model.add( Conv2D(64,(3,3),input_shape=X.shape[1:],activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add( Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add( Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
#model.add(Dropout(1.0))

model.add(Flatten())

model.add(Dense(45,activation='softmax'))
#model.add(Dense(41,activation='softmax'))
```

```python

model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
import matplotlib.pyplot as plt

history=model.fit(X,Y,epochs=550,validation_split=0.1)
print(history.history.keys())

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```
